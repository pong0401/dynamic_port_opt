# Latest Tactical TH/Gold/BTC Last Weights

This notebook refreshes the latest effective weights for `Final Best Sharpe Tactical TH/Gold/BTC 65/25/10 Gold crash protection`.

It intentionally avoids the large `us_th_tactical_perf_momentum` notebook gate and does not run the tactical/sweep notebook workflow. The refresh script downloads only a rolling fresh window for US stocks, Thailand stocks, Gold, BTC, FX, SPY, VIX, and SET, then recomputes the latest PIT sleeve weights and asset-level daily exposure.

```python
from pathlib import Path
import importlib
import sys

import pandas as pd

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 200)
pd.set_option("display.float_format", lambda value: f"{value:,.6f}")

ROOT = Path.cwd().resolve()
while not (ROOT / "scripts" / "recheck_us_th_tactical_gold_btc_latest_weights.py").exists() and ROOT.parent != ROOT:
    ROOT = ROOT.parent

SCRIPTS = ROOT / "scripts"
SRC = ROOT / "src"
for path in [SCRIPTS, SRC]:
    if str(path) not in sys.path:
        sys.path.insert(0, str(path))

import recheck_us_th_tactical_gold_btc_latest_weights as latest_weights
latest_weights = importlib.reload(latest_weights)

RUN_FRESH_REFRESH = True

if RUN_FRESH_REFRESH:
    latest_weights.main()

RESULT_PREFIX = latest_weights.RESULT_PREFIX
RESULT_DIR = ROOT / "result"
security_path = RESULT_DIR / f"{RESULT_PREFIX}_latest_effective_security_weights_thb.csv"

security_weights = pd.read_csv(security_path)

print(f"Loaded latest outputs from: {RESULT_DIR}")
```

## Last Weights

```python
last_weight = (
    security_weights
    .assign(
        weight=security_weights["Effective Weight"],
        exposure=security_weights["Daily Exposure"],
    )
    .rename(columns={"Asset": "asset"})
    .loc[:, ["asset", "weight", "exposure"]]
    .sort_values("weight", ascending=False)
    .reset_index(drop=True)
)
last_weight
```
